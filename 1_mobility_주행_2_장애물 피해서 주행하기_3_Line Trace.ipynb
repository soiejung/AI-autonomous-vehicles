```python
from RobokitRS.RobokitRS import *
import time as t
rs = RobokitRS()
```

```python
rs.port_open("COM7")
```

```python
def forward():
    rs.motor_write(0,1,3.1)
    rs.motor_write(1,0,2)
    rs.motor_write(2,1,3.2)
    rs.motor_write(3,0,2)

    
    
def stop():
    rs.motor_write(0,1,0)
    rs.motor_write(1,0,0)
    rs.motor_write(2,1,0)
    rs.motor_write(3,0,0)
    t.sleep(0.1)
    
def reverse():
    rs.motor_write(0,0,10.2)
    rs.motor_write(1,1,10)
    rs.motor_write(2,0,10.2)
    rs.motor_write(3,1,10)
    
def right(num):
    rs.motor_write(0,1,10.2)
    rs.motor_write(1,1,10)
    rs.motor_write(2,0,10.2)
    rs.motor_write(3,0,10)
    t.sleep(num)
    
def left(num):
    rs.motor_write(0,0,10.2)
    rs.motor_write(1,0,10)
    rs.motor_write(2,1,10.2)
    rs.motor_write(3,1,10)
    t.sleep(num)
    
def left_turn(num):
    rs.motor_write(0,0,10)
    rs.motor_write(1,0,10.2)
    rs.motor_write(2,0,10)
    rs.motor_write(3,0,10.2)
    t.sleep(num)
    
def right_turn(num):
    rs.motor_write(0,1,10.2)
    rs.motor_write(1,1,10)
    rs.motor_write(2,1,10.2)
    rs.motor_write(3,1,10)
    t.sleep(num)
    
def front_left(num):
    rs.motor_write(1,0,10)
    rs.motor_write(2,1,10.2) 
    t.sleep(num)
    
def front_right(num):
    rs.motor_write(0,1,10.2)
    rs.motor_write(3,0,10)
    t.sleep(num)
    
def L_turn():
    rs.motor_write(0,0,0)
    rs.motor_write(1,0,15)
    rs.motor_write(2,0,0)
    rs.motor_write(3,0,15)
    
    
def R_turn():
    rs.motor_write(0,1,15)
    rs.motor_write(1,1,0)
    rs.motor_write(2,1,15)
    rs.motor_write(3,1,0)
    
```

```python
#초음파 센서 감지
rs.sonar_begin(13)

i=0
while i<10:
    SONAL = rs.sonar_read(13)
    t.sleep(1)
    print(SONAL)
    i=i+1
```

```python
#1. RGB LED 색상 설정 후 ON
rs.set_rgb_led_color(12,100,100,100)
rs.set_rgb_led_on(12)

#2. RGB LED 색상 바꾸기
rs.set_rgb_led_red(12)
rs.delay()
rs.set_rgb_led_on(12)
rs.delay()
rs.set_rgb_led_orange(12)
rs.delay()
rs.set_rgb_led_off(12)
```

```python
#거리에 따른 색상 변경

rs.sonar_begin(13)

i=0
while True:
    SONAL = rs.sonar_read(13)
    t.sleep(0.1)
    print(SONAL)
    
    if 40< SONAL:
        rs.set_rgb_led_color(12,100,100,100)
        rs.set_rgb_led_on(12)
        
    elif 35< SONAL:
        rs.set_rgb_led_color(12,100,0,0)
        rs.set_rgb_led_on(12)
        
    elif 30< SONAL:
        rs.set_rgb_led_color(12,0,0,100)
        rs.set_rgb_led_on(12)
        
    elif 25< SONAL:
        rs.set_rgb_led_color(12,0,100,0)
        rs.set_rgb_led_on(12)
        
    elif 20< SONAL:
        rs.set_rgb_led_color(12,0,100,100)
        rs.set_rgb_led_on(12)
        
    elif 15< SONAL:
        rs.set_rgb_led_color(12,100,0,100)
        rs.set_rgb_led_on(12)
        
    elif 10<SONAL:
        rs.set_rgb_led_color(12,100,100,0)
        rs.set_rgb_led_on(12)
        
    elif 5<SONAL:
        rs.set_rgb_led_color(12,50,100,100)
        rs.set_rgb_led_on(12)
        
    else :
        rs.set_rgb_led_off(12)
        
        
```

```python
rs.sonar_begin(13)
i = 0
num=4.3
try:
    while True:
        S = rs.sonar_read(13)  
        print(S)
        t.sleep(0.1)
        if i >= 3:
            break
        
        else :
        
                if S < 15:
                    stop()
                    t.sleep(0.5)
                    i+=1
                
                    if i==1:
                        rs.set_rgb_led_color(12,100,0,0)
                        rs.set_rgb_led_on(12)
                        num=2
                
                    if i==2:
                        rs.set_rgb_led_color(12,0,100,0)
                        rs.set_rgb_led_on(12)
                        num=1
                
                    if i==3:
                        rs.set_rgb_led_color(12,0,0,100)
                        rs.set_rgb_led_on(12)
                        num=1.8
            
                    right_turn(2)
                    rs.set_rgb_led_off(12)
            
        forward(num)

    stop()

except:
    stop()
```
